# Data Analysis Part 1 Pandas Assignment

```python
import pandas as pd
import glob
path = "./*.csv"
li=[]
for fname in glob.glob(path):
    df=pd.read_csv(fname , index_col=None , header=0)
    li.append(df)
us_census=pd.concat(li , axis=0,ignore_index=True)
us_census.head()
```

```python
import numpy as np
```

```python
us_census.dtypes
```

```python
new_us_census=us_census.drop(columns=['Unnamed: 0','location', 'product_type','product_description','quantity','price'] ,
                          inplace=True)
```

```python
us_census.head()
```

```python
us_census=us_census.dropna()
```

```python
us_census[['Male' , 'Female']]=us_census.GenderPop.str.split('_',expand=True)
```

```python
us_census.head()
```

```python
new_us_census
```

```python
us_census['Income']=us_census.Income.str.strip('$')
```

```python
us_census=us_census.replace('%', '',regex=True)
```

```python
us_census=us_census.replace('M', '',regex=True)
```

```python
us_census=us_census.replace('F', '',regex=True)
```

```python
us_census
```

```python
us_census.drop(columns=['GenderPop'] , inplace=True)
```

```python
us_census.dtypes
```

```python
us_census.loc[:,'Hispanic':'Income']=round(us_census.loc[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
us_census
```

```python
us_census['Male']=us_census['Male'].astype(int)
```

```python
np.where(us_census.applymap(lambda x: x==''))
```

```python
us_census.drop(columns=['Female'], inplace=True)
```

```python
us_census['Female']=us_census['TotalPop']-us_census['Male']
```

```python
us_census
```

```python
us_census.dtypes
```

```python
import matplotlib.pyplot as plt
plt.scatter(us_census['Female'], us_census['Income'])
plt.xlabel('Salary $')
plt.ylabel('Production')
plt.show()
```

```python
us_census.duplicated().any()
```

```python
us_census.drop_duplicates(inplace=True)
```

```python
plt.scatter(us_census['Female'], us_census['Income'])
plt.xlabel('Salary $')
plt.ylabel('Production')
plt.show()
```
